### Karatsuba Algorithm

*Multiplication with Recursive function.*

* Flow of the Algorithm:
  1. Input two numbers. 
  2. Check if the length of either of the number is less then 10 i.e., if any of the number is of single digit. If yes then we can perform direct multiplication. This will be our base case of the the recursive function. 
  3. If the numbers are not single digits, then we start with the Karatsuba algorithm. 
  4. Find the maximum number of digits from eith of the number.
  5. Check if the max number of digits is an even or odd number. 
     * If it is an odd number then you will have to reduce the number by one and then divide the result by two and save it in another variable. 
     * If it is an even number then you directly divide the number by two and save it in another variable.
  6. The number generated after dividing the max length of the either number by two will be helpful while padding the individual results with zeros. The of check will help to compensate for the problem when trying to divide an odd number with two. 
  7. Now you can form the upper and lower halfs of both the numbers by taking the division of the number by 10^(max length of wither number/two) and the modules of the number with 10^(max length of wither number/two). Repeat this step for both the numbers. 
  8. Now you have a, b, c, and d. Compute ac, cd and ad + cd. Resursively call the function and pass the values of a, b, c, and d generated in the previous step. 
  9. At the end of recursion you will have your product of two large numbers. 


* Things to keep in mind:
  1. The max length of either numbers must be chaecked for odd and even. If it is odd then do convert it to even because padding is done in multiples of two.
  2. len() finction doesnot work on int. Take care of the data type and convert int to str and vice vera when required. 

```python
'''This cell will show you set by step computation of the algorithm.
Next cell will compute the runtime to execute the algorithm.
Also, code in the next cell is without extra print statements.'''

iteration = 0
def karatsuba(x,y): 
    global iteration
    # base case
    if x < 10 or y < 10: # alternative way to say if it is a single digit number. 
        print(f'Reached the base case. Multiplying {x}*{y} = {x*y}.')
        print('Returning the result...')
        print()
        return x*y    
    else:
        print(f'Check which number has more digits: {x} or {y} and save the maximum length')
        n = max(len(str(x)), len(str(y)))
        print(f'The max length of either number is {n}')
        if n % 2 != 0: # check if length is odd or even. 
            print(f'Since {n} is an odd number we reduce its value by one to over come the padding problem for ad_bc term, i.e., ad + bd.')
            n -= 1
        print('Now since the max length is an even number we divide it by two.')
        print(f'This will be used for padding and generating upper and lower halfs of {x} and {y}')
        print(f'a and b for {x} and c and d for {y}')
        nbytwo = int(n/2)

        a, b = divmod(x, 10**nbytwo) # get the upper and lower halfs of x in a and b.
        print(f'Upper half for {x}: {a}')
        print(f'Lower half for {x}: {b}')
        c, d = divmod(y, 10**nbytwo) # get the upper and lower halfs of y in c and d.
        print(f'Upper half for {y}: {c}')
        print(f'Lower half for {y}: {d}')
        iteration += +1
        print(f'We recursively call the function {iteration} times...')
        print()
        ac = karatsuba(a,c) # recursive call to the function 
        bd = karatsuba(b,d)
        ad_bc = karatsuba((a+b),(c+d)) - ac - bd
        
        # padding ac with the the max length of either numbers 10**n. Here n can be odd or even. 
        # If n was odd, then that would be a problem for padding ad_bc because the pads have to 
        # be half of ac. Thus, the check foor odd and even resolves the issue. 
        print()
        return ((ac*(10**n)) + ((ad_bc)*(10**nbytwo)) + bd)
        

x = int(input('Enter first number: '))
y = int(input('Enter second number: ')) 
print()

print(f'Final answer is: {karatsuba(x,y)}')
print(f'After iterating {iteration} times.')
```

```python
import time
def karatsuba(x,y):    
    # base case
    if x < 10 or y < 10: # alternative way to say if it is a single digit number. 
        return x*y    
    else:
        n = max(len(str(x)), len(str(y)))
        if n % 2 != 0: # check if length is odd or even. 
            n -= 1
        nbytwo = int(n/2)

        a, b = divmod(x, 10**nbytwo) # get the upper and lower halfs of x in a and b.
        c, d = divmod(y, 10**nbytwo) # get the upper and lower halfs of y in c and d.

        ac = karatsuba(a,c) # recursive call to the function 
        bd = karatsuba(b,d)
        ad_bc = karatsuba((a+b),(c+d)) - ac - bd
        
        # padding ac with the the max length of either numbers 10**n. Here n can be odd or even. 
        # If n was odd, then that would be a problem for padding ad_bc because the pads have to 
        # be half of ac. Thus, the check foor odd and even resolves the issue. 
        return ((ac*(10**n)) + ((ad_bc)*(10**nbytwo)) + bd)
    
x = int(input('Enter first number: '))
y = int(input('Enter second number: ')) 
start = time.time()
print ('Answer: ', karatsuba(x,y))
print('Execution time: ', time.time()-start)
```
