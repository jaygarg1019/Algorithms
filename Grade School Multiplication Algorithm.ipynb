### Grade School Multiplication Algorithm
*Build from scratch*

* Flow of the Algorithm:
  1. Input two numbers. 
  2. For each digit of the second number, multiply with every digit of the first number. 
  3. Keep track of the carry and generate a result after fished multiplying one digit from second number to every digit of the first number. 
  4. Keep adding the result.
     * For consecutive results keep incrementing the number of zeros trailing each result. This is taken care by multiplying the result with 10^(the units place of the digit, eg. tens[1], hundred[2], thousand[3])
  5. Finally print the result. 


* Things to keep in mind:
  1. The multiplication starts from the right most digit. List indexing in Python starts from left most as zero. Reverse the list order. 
  2. If the multiplication is between one digit on the second number and the last digit of the first number, the carry should not be generated. Instead the complete answer should be saved together. Create a special case which checks if the digit is are the last place or not. 


```python
'''This cell will show you set by step computation of the algorithm.
Next cell will compute the runtime to execute the algorithm.
Also, code in the next cell is without extra print statements.'''

n = list(map(int,list(input('Enter frist number:'))))
m = list(map(int,list(input('Enter second number:'))))
print()

print('This is how the numbers are present in the list: ',n,' and ', m)
n.reverse()
m.reverse()
print(f'After reversing first number: {n} and second number: {m}')
print()

result = [] 
carry = 0
      
for i in range(len(m)):
    print('-------------------------------------------')
    print(f'For {i+1}st digit in second number: ')
    for j in range(len(n)):
        print(m[i], '*', n[j], '= ', end= '')
        ans = list(map(int,list(str((m[i]*n[j])+carry))))
        print(''.join(list(map(str,ans))), end='')
        if len(ans) == 2 and j != len(n)-1:
            carry = ans[0]
            result.insert(0,ans[1])
            print(' carry = ', carry, ' saved digit = ',''.join(list(map(str,result))))
        elif len(ans) == 2 and j == len(n)-1:
            carry = 0
            result.insert(0,ans[1])
            result.insert(0,ans[0])
            print(' carry = ', carry, ' saved digit = ',''.join(list(map(str,result))))
        else:
            carry = 0
            result.insert(0,ans[0])
            print(' carry = ', carry, ' saved digit = ',''.join(list(map(str,result))))
        ans.clear()
    if i == 0:
        previous_result = int(''.join(map(str,result)))
        final_result = previous_result
    else:
        final_result = int(''.join(map(str,result)))*(10**i) + previous_result  
        previous_result = final_result
    print('result = ', previous_result)
    result.clear()
    
print()
print('***************************************************')
print('Final Answer = ', final_result)
print('***************************************************')
```

```python
import time 
n = list(map(int,list(input('Enter frist number:')))) # convert input string of numbers to list of numbers. 
m = list(map(int,list(input('Enter second number:'))))
start = time.time() # start timer 
n.reverse()
m.reverse()
result = [] 
carry = 0
      
for i in range(len(m)):
    for j in range(len(n)):
        ans = list(map(int,list(str((m[i]*n[j])+carry))))
        if len(ans) == 2 and j != len(n)-1:      # if the ans has a carry and the multiplied digit is not in the last place.
            carry = ans[0]
            result.insert(0,ans[1])
        elif len(ans) == 2 and j == len(n)-1:    # if the ans has a carry and the multiplied digit is at the last place. 
            carry = 0
            result.insert(0,ans[1])
            result.insert(0,ans[0])
        else:                                   # if the ans doesn't have a carry.  
            carry = 0
            result.insert(0,ans[0])
        ans.clear()
    if i == 0:
        previous_result = int(''.join(map(str,result))) # convert list of numbers to a single number, integer. 
        final_result = previous_result
    else:
        final_result = int(''.join(map(str,result)))*(10**i) + previous_result # increment trailing zeros to the result.
        previous_result = final_result
    result.clear()
    
print('Answer:', final_result)
print('Execution time: ', time.time()-start) # print the time taken to execute the algo. 
```
